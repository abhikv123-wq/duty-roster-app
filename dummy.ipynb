```python
from datetime import datetime, timedelta

def generate_shift_roster(start_date):
    # Define the shift configuration
    shift_config = {
        'A': ['Evening', 'Evening', 'Off', 'Off', 'Morning', 'Morning', 'Night', 'Night'],
        'B': ['Morning', 'Morning', 'Off', 'Off', 'Evening', 'Evening', 'Night', 'Night'],
        'C': ['Night', 'Night', 'Off', 'Off', 'Morning', 'Morning', 'Evening', 'Evening'],
        'D': ['Off', 'Off', 'Evening', 'Evening', 'Morning', 'Morning', 'Night', 'Night']
    }

    # Initialize the shift roster
    shift_roster = {}

    # Generate the shift roster for each group
    for group, shifts in shift_config.items():
        shift_roster[group] = []
        for i in range(len(shifts)):
            shift_roster[group].append((start_date + timedelta(days=i), shifts[i]))

    return shift_roster

def get_shift_for_date(shift_roster, date):
    date = datetime.strptime(date, '%d/%m/%Y')
    for group, shifts in shift_roster.items():
        for shift in shifts:
            if shift[0] == date:
                return group, shift[1]
    return None, None

def main():
    start_date = datetime.strptime('31/12/2024', '%d/%m/%Y')
    shift_roster = generate_shift_roster(start_date)

    print("Shift Roster:")
    for group, shifts in shift_roster.items():
        print(f"Group {group}:")
        for shift in shifts:
            print(f"{shift[0].strftime('%d/%m/%Y')}: {shift[1]}")

    while True:
        date = input("Enter a date (dd/mm/yyyy) to get the corresponding shift: ")
        group, shift = get_shift_for_date(shift_roster, date)
        if group and shift:
            print(f"Group {group} has a {shift} shift on {date}.")
        else:
            print("No shift found for the given date.")

if __name__ == "__main__":
    main()
```
